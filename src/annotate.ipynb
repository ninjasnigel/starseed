Randomize variables

```python
from random import randint
from script.annotation import region_to_string, convert_to_region
SIZE = 800
X_START = randint(SIZE*2, 120000-SIZE*2)
Y_START = randint(3000, 6000)
PATH = '../data/together_annotation'
box = [(X_START, Y_START, X_START+SIZE, Y_START+SIZE)]
NAME = region_to_string(convert_to_region(box)[0])
```

Annotate

```python
from script.annotation import annotate_from_fits, regions_to_regfile, convert_to_region

boxes = annotate_from_fits(X_START, Y_START, SIZE, PATH, NAME)

regions = convert_to_region(boxes)

# Save annotations to regfile
if boxes: regions_to_regfile(boxes, NAME, PATH)
else:
    import os
    if os.path.exists(f"{PATH}/annimg.png"):
        os.remove(f"{PATH}/annimg.png")

if os.path.exists(f"{PATH}/{NAME}.png"):
    os.remove(f"{PATH}/{NAME}.png")
```
